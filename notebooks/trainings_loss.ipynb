```python
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize=(8, 4), dpi = 150)

df = pd.read_csv("../output/TTBar_600000_training_log.csv")

ax.plot(df["epoch"][2:], df["train_loss"][2:], label="train")
ax.plot(df["epoch"][2:], df["val_loss"][2:], label="val")

ax.legend()

#ax.set_yscale("log")
ax.set_title("Training/Validation Loss")

ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
plt.show()
```
